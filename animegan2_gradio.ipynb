# ライブラリのインポート

```python
import gradio as gr
from PIL import Image
import cv2
import torch
import numpy as np
```

# モデルのロード

```python
model = torch.hub.load("bryandlee/animegan2-pytorch:main", "generator", pretrained="celeba_distill")
face2paint = torch.hub.load("bryandlee/animegan2-pytorch:main", "face2paint")
```

# 推論用関数

```python
def animegan2_inference(input_img):
    input_img = input_img.convert("RGB")
    img_np = np.array(input_img)
    img_cv = cv2.cvtColor(img_np, cv2.COLOR_RGB2BGR)
    img_for_model = Image.fromarray(cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB))
    output_img = face2paint(model, img_for_model)
    return output_img
```

# インターフェースの表示

```python
iface = gr.Interface(
    fn=animegan2_inference,
    inputs=gr.Image(type="pil"),
    outputs=gr.Image(type="pil"),
    title="アニメ風画像を作ろう"
)

iface.launch()
```
